# 1. Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the raw CSV 
# Expected local path = data/raw/india_housing_prices.csv
DATA_PATH = r"data/raw/india_housing_prices.csv"

df = pd.read_csv(DATA_PATH)
# Normalize blank / whitespace only strings to NA for consistent cleaning
df = df.replace(r"^\s*$", pd.NA, regex=True)

# Sanity checks = sample rows, schema, and summary stats
display(df.head())
df.info()
df.describe(include="all").T
```

### 1.1 Target distribution + numeric summary

- **Aim**: understand the spread of `Price_in_Lakhs` and basic numeric ranges.
- **Inference**: if simple linear patterns are weak, non-linear ensembles/AutoML are a reasonable default.


```python
# Target and a small numeric subset for quick EDA
target = "Price_in_Lakhs"
numeric_cols = [
    "BHK", "Size_in_SqFt", "Price_per_SqFt", "Year_Built",
    "Floor_No", "Total_Floors", "Age_of_Property",
    "Nearby_Schools", "Nearby_Hospitals",
]

# Target distribution (helps choose robust metrics / models)
sns.histplot(df[target], kde=True)
plt.title("Distribution of Price_in_Lakhs")
plt.show()

# Numeric ranges and summary stats
df[numeric_cols].describe().T
```

### 1.2 Missingness

- **Aim**: check whether we need imputation and where
- **Inference**: near zero missingness means imputation is just a safety net 


```python
missing_pct = df.isna().mean().sort_values(ascending=False)
display(missing_pct.to_frame("missing_fraction").style.bar(color="lightcoral"))

sns.heatmap(df[numeric_cols].isna(), cbar=False)
plt.title("Missing pattern (numeric subset)")
plt.show()
```

### 1.3 Categorical counts

- **Aim**: spot rare categories, unexpected labels, and balance issues
- **Inference**: heavily imbalanced categories may need grouping, otherwise one hot is fine


```python
categorical_cols = [c for c in df.columns if df[c].dtype == "object" and c not in ("Amenities", target)]
for col in categorical_cols:
    print(f"\n--- {col} ---")
    display(df[col].value_counts(dropna=False).head(10))
```

### 1.4 Correlation 

- **Aim**: see whether a simple linear signal exists
- **Inference**: weak linear correlations and diffuse pairplots suggest non linear structure = ensembles / AutoML friendly


```python
corr = df[numeric_cols + [target]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation matrix")
plt.show()

sns.pairplot(df[numeric_cols + [target]].sample(500, random_state=42), diag_kind="kde")
plt.suptitle("Pairplot (sampled)", y=1.02)
plt.show()
```

### 1.5 Median price pivot

- **Aim**: show summary of price differences across `Property_Type` Ã— `Furnished_Status`
- **Inference**: clear gaps here indicate categorical effects the model should capture


```python
pivot = (
    df.pivot_table(
        values=target,
        index="Property_Type",
        columns="Furnished_Status",
        aggfunc="median"
    )
)
display(pivot)
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Median price by property and furnishing")
plt.show()
```

# 3. Autosklearn House Price Prediction Modeling
This notebook implements:
1. Data loading and preprocessing using the project pipeline
2. Autosklearn regression modeling
3. Baseline comparison with Random Forest and XGBoost
4. Performance evaluation and residual analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time

# Scikitlearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# XGBoost
import xgboost as xgb

# Autosklearn
import autosklearn.regression
from autosklearn.regression import AutoSklearnRegressor

# Model persistence
import joblib

# Project wrapper layer 
from auto_sklearn_utils import prepare_data, load_housing_data, to_float32
```

### 3.1 Load and preprocess (shared pipeline)

- **Aim**: generate `X_train`, `X_test`, `y_train`, `y_test` once using the project pipeline
- **Inference**: downstream models should differ only by learning algorithm not by preprocessing as it should remain the same


```python
DATA_PATH = "data/raw/india_housing_prices.csv"

# To produce train/test splits and processed feature matrices
X_train, X_test, y_train, y_test, preprocessor, feature_names = prepare_data(
    data_path=DATA_PATH,
    test_size=0.2,
    random_state=42,
)

print(f"Training set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
```

### 3.1 Autosklearn Model

```python
# Ensure feature matrices are float32 (faster)
# Implemented as a reusable helper in auto_sklearn_utils.py
X_train = to_float32(X_train)
X_test  = to_float32(X_test)

print("X_train dtype:", getattr(X_train, "dtype", None), "shape:", getattr(X_train, "shape", None))
print("X_test  dtype:", getattr(X_test, "dtype", None), "shape:", getattr(X_test, "shape", None))  
```

```python
# AutoML configuration
# - per_run_time_limit may be capped if it exceeds the total time budget
# - memory_limit is in MB
# - tmp_folder isolates artifacts per run
automl = AutoSklearnRegressor(
    time_left_for_this_task=900,
    per_run_time_limit=900,
    n_jobs=1,
    seed=42,
    memory_limit=10240,
    ensemble_kwargs={"ensemble_size": 10},
    resampling_strategy="holdout",
    include={
        "regressor": [
            "random_forest",
            "gradient_boosting",
            "extra_trees",
            "adaboost",
            "decision_tree",
        ],
    },
    tmp_folder=f"/tmp/autosklearn_run_{int(time.time())}",
)

print("Training auto-sklearn regressor")
# Subsample for a faster demo run
# Increase/remove for best quality (more compute needed)
automl.fit(X_train[:20000], y_train[:20000])

print("\n" + "=" * 60)
print("AUTO-SKLEARN TRAINING SUMMARY")
print("=" * 60)
print(automl.sprint_statistics())
```

### 3.2 Autosklearn inspection 

- **Aim**: peek into the ensemble composition and the search results (`cv_results_`) to understand which model families performed well
- **Inference**: if the ensemble heavily weights one family thatâ€™s a clue about what works for this dataset


```python
# inspect the ensemble composition
print("ENSEMBLE LEADERBOARD")

print(automl.leaderboard())

print("\n" + "="*60)
print("MODELS IN ENSEMBLE")
print("="*60)
print(automl.show_models())

# get detailed model weights
print("ENSEMBLE WEIGHTS")
models_with_weights = automl.get_models_with_weights()
for weight, model in models_with_weights:
    print(f"\nWeight: {weight:.4f}")
    print(f"Model: {model}")
    print("-"*60)
```

```python
# Analyze autosklearn run results 

print("AUTO-SKLEARN RUN RESULTS ANALYSIS")

# Cv_results_ is dictlike so we convert first
cv_df = pd.DataFrame(automl.cv_results_)
print(f"Total configurations recorded: {len(cv_df)}")
print("Available columns:", cv_df.columns.tolist())

# Pick a status column if present
status_col = None
for c in ["status", "status__", "run_status", "run_status__"]:
    if c in cv_df.columns:
        status_col = c
        break

if status_col is not None:
    print(f"\nStatus breakdown ({status_col}):")
    print(cv_df[status_col].value_counts(dropna=False).to_string())
else:
    print("\nNo status column found in cv_results_ (version-dependent).")

# Pick a score column 
score_col = None
for c in ["mean_test_score", "cost", "score", "mean_validation_score", "mean_cross_validation_score"]:
    if c in cv_df.columns:
        score_col = c
        break

if score_col is not None:
    print(f"\nPerformance statistics ({score_col}):")
    s = cv_df[score_col]
    s_num = pd.to_numeric(s, errors="coerce")
    print(f"Best:  {s_num.max():.6f}")
    print(f"Mean:  {s_num.mean():.6f}")
    print(f"Std:   {s_num.std():.6f}")
    print(f"Worst: {s_num.min():.6f}")
else:
    print("\nNo recognizable score column found in cv_results_.")

# Regressor choice column 
choice_col = None
for c in ["param_regressor:__choice__", "param_regressor__choice__", "regressor", "model"]:
    if c in cv_df.columns:
        choice_col = c
        break

# Show top 5 (if rank + score exist)
rank_col = None
for c in ["rank_test_scores", "rank_test_score", "rank"]:
    if c in cv_df.columns:
        rank_col = c
        break

print("\nTOP 5 CONFIGURATIONS (if available)")
if rank_col is not None and score_col is not None:
    cols = [rank_col, score_col] + ([choice_col] if choice_col is not None else [])
    top = cv_df.sort_values(rank_col).head(5)[cols]
    print(top.to_string(index=False))
else:
    print("Rank/score columns not available for a clean top-5 table.")

# Plots
plt.figure(figsize=(12, 5))

if score_col is not None:
    plt.subplot(1, 2, 1)
    s_num = pd.to_numeric(cv_df[score_col], errors="coerce").dropna()
    plt.hist(s_num, bins=20, edgecolor="black")
    plt.xlabel(score_col)
    plt.ylabel("Frequency")
    plt.title("Distribution of model scores")
    plt.grid(True, alpha=0.3)

if choice_col is not None:
    plt.subplot(1, 2, 2)
    counts = cv_df[choice_col].astype(str).value_counts()
    counts.plot(kind="barh")
    plt.xlabel("Count")
    plt.ylabel("Regressor type")
    plt.title("Models tried by type")

plt.tight_layout()
plt.show()
```

### 3.3 Baseline models

- **Aim**: compare AutoML to strong widely used tabular regressors
- **Premise**: baselines ensure AutoML gains are real 

```python
# random forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
print("Training Random Forest")
rf_model.fit(X_train[:20000], y_train[:20000])

# xgboost
xgb_model = xgb.XGBRegressor(n_estimators=100, random_state=42, n_jobs=-1)
print("Training XGBoost")
xgb_model.fit(X_train[:20000], y_train[:20000])
```

### 3.4 Evaluation & comparison

- **Aim**: evaluate AutoML and baselines on the same test set
- **Metrics**:
  - **MAE**: easier to interpret in the targetâ€™s units (lakhs)
  - **RMSE**: highlights occasional large misses


```python
# Build results_df + predictions 
models = {
    "Random Forest": rf_model,
    "XGBoost": xgb_model,
    "Auto-sklearn": automl,
}

predictions = {}
rows = []

for name, model in models.items():
    y_pred = model.predict(X_test)
    predictions[name] = y_pred

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    rows.append({"Model": name, "MAE": mae, "RMSE": rmse})

results_df = pd.DataFrame(rows).sort_values("MAE").reset_index(drop=True)
results_df

# Performance improvement over baseline
rf_mae = results_df[results_df['Model'] == 'Random Forest']['MAE'].values[0]
xgb_mae = results_df[results_df['Model'] == 'XGBoost']['MAE'].values[0]
auto_mae = results_df[results_df['Model'] == 'Auto-sklearn']['MAE'].values[0]

print("="*60)
print("BASELINE COMPARISON")
print("="*60)

print(f"Random Forest MAE:{rf_mae:.4f}")
print(f"XGBoost MAE:{xgb_mae:.4f}")
print(f"Auto sklearn MAE:{auto_mae:.4f}")

best_baseline = min(rf_mae, xgb_mae)
improvement = ((best_baseline - auto_mae) / best_baseline) * 100

if improvement > 0:
    print(f"\nAutosklearn outperforms best baseline by {improvement:.2f}%")
else:
    print(f"\nAutosklearn is within {abs(improvement):.2f}% of best baseline")
```

```python
# Visualization of metrics
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.barplot(data=results_df, x="Model", y="MAE", hue="Model")
plt.title("MAE Comparison")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(data=results_df, x="Model", y="RMSE", hue="Model")
plt.title("RMSE Comparison")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Detailed autosklearn performance insights
print("AUTO SKLEARN DETAILED PERFORMANCE ANALYSIS")

# Compare autosklearn ensemble vs single best model in ensemble
models_in_ensemble = automl.get_models_with_weights()
if models_in_ensemble:
    print(f"\nEnsemble composition:")
    print(f"Number of models in ensemble:{len(models_in_ensemble)}")
    print(f"Total weight:{sum(w for w, _ in models_in_ensemble):.4f}")
    
    # get predictions from ensemble
    ensemble_pred = automl.predict(X_test)
    ensemble_mae = mean_absolute_error(y_test, ensemble_pred)
    ensemble_rmse = np.sqrt(mean_squared_error(y_test, ensemble_pred))
    ensemble_r2 = automl.score(X_test, y_test)
    
    print(f"\nEnsemble performance:")
    print(f"MAE:{ensemble_mae:.4f}")
    print(f"RMSE:{ensemble_rmse:.4f}")
    print(f"RÂ²:{ensemble_r2:.4f}")
    
    # Compare with best single model in ensemble
    print(f"\nTop 3 models by weight:")
    for i, (weight, model) in enumerate(models_in_ensemble[:3], 1):
        print(f"  {i}. Weight: {weight:.4f} | {str(model)[:80]}...")

# Performance improvement over baseline
rf_mae = results_df[results_df['Model'] == 'Random Forest']['MAE'].values[0]
xgb_mae = results_df[results_df['Model'] == 'XGBoost']['MAE'].values[0]
auto_mae = results_df[results_df['Model'] == 'Auto-sklearn']['MAE'].values[0]

print("\nPERFORMANCE COMPARISON")
print(f"Random Forest MAE:{rf_mae:.4f}")
print(f"XGBoost MAE:{xgb_mae:.4f}")
print(f"Auto sklearn MAE:{auto_mae:.4f}")

best_baseline = min(rf_mae, xgb_mae)
improvement = ((best_baseline - auto_mae) / best_baseline) * 100

if improvement > 0:
    print(f"\nAutosklearn outperforms best baseline by {improvement:.2f}%")
else:
    print(f"\nAutosklearn is within {abs(improvement):.2f}% of best baseline")
```

### 3.5 Residual analysis (best model)

- **Aim**: diagnose error patterns beyond a single score
- **Inference**:
  - predicted vs actual should align near the diagonal
  - residual distribution should be centered near 0 (large outliers indicate potential issues)


```python
# select best model based on MAE
best_model_name = results_df.sort_values("MAE").iloc[0]["Model"]
best_preds = predictions[best_model_name]

residuals = y_test - best_preds

plt.figure(figsize=(12, 5))

# predicted vs actual
plt.subplot(1, 2, 1)
plt.scatter(y_test, best_preds, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(f"{best_model_name}: Predicted vs Actual")

# residual distribution
plt.subplot(1, 2, 2)
sns.histplot(residuals, kde=True)
plt.title(f"{best_model_name}: Residual Distribution")
plt.xlabel("Error (Actual - Predicted)")

plt.tight_layout()
plt.show()
```
