```python
import pandas as pd

PATH_IN  = "Sentiment140_raw.csv"           
PATH_OUT = "Sentiment140_raw_fixed.csv"    

df = pd.read_csv(PATH_IN)

expected = [
    "target",        
    "label_str",    
    "id",            
    "date",          
    "flag",          
    "user",          
    "text",          
    "text_clean",    
    "char_count",    
    "word_count"     
]

assert len(df.columns) == len(expected), f"Expected 10 columns, found {len(df.columns)}"

df.columns = expected

if "character_count" in df.columns:
    df = df.rename(columns={"character_count": "char_count"})

df.to_csv(PATH_OUT, index=False)
print("Wrote:", PATH_OUT)
print("Columns now:", list(df.columns))
```
